```python
from scikit_quri.circuit import create_qcl_ansatz
from quri_parts_scaluq.circuit import convert_parametric_circuit

# import pandas as pd
# from sklearn import datasets
# from sklearn.model_selection import train_test_split

# iris = datasets.load_iris()
# df = pd.DataFrame(iris.data, columns=iris.feature_names)
# x = df.loc[:, ["petal length (cm)", "petal width (cm)"]]
# x_train, x_test, y_train, y_test = train_test_split(x, iris.target, test_size=0.25, random_state=0)
# x_train = x_train.to_numpy()


n_qubits = 5
circuit = create_qcl_ansatz(n_qubits, 3, 1.0)
print(circuit.parameter_count)

converted_circuit, param_mapper = convert_parametric_circuit(circuit.circuit)
```

```python
from collections.abc import Sequence, Mapping


def batched_param_mapper(
    params_list: Sequence[Sequence[float]],
) -> dict[str, list[float]]:
    mapped = [param_mapper(params) for params in params_list]
    keys = mapped[0].keys()
    return {k: [m[k][0] for m in mapped] for k in keys}
```

```python
from scaluq.default.f64 import StateVectorBatched

state_vector = StateVectorBatched(3, n_qubits)
state_vector.set_zero_state()
init_param = [0.1] * circuit.parameter_count
init_params2 = [0.2] * circuit.parameter_count
converted_circuit.update_quantum_state(state_vector, batched_param_mapper([init_param, init_params2]))
print(batched_param_mapper([init_param, init_params2]))
```
